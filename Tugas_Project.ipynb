```python
#daftar library yang digunakan
import pandas as pd
from datetime import datetime #memanggil modul waktu
import pytz #memanggil modul zona waktu
import random
```

```python
#global function dan variabel
def pembukaan():
    #menampilkan pesan awal dan tanggal saat transaksi dibuat
    waktu = datetime.now(pytz.timezone('Asia/Jakarta')) #mengambil data waktu Jakarta saat program digunakan
    teks_pembukaan ='Selamat Datang di Aplikasi Super Cashier'
    print('-'*len(teks_pembukaan))
    print(f'{teks_pembukaan}\n{waktu.strftime("%a, %d %B %Y - %H:%M:%S")}') #menampilkan tanggal, bulan, tahun, dan waktu
    print('-'*len(teks_pembukaan))


def buat_transaksi():
    while True:
        try:
            pilihan_menu_awal = input(f'Apakah Anda ingin membuat transaksi baru? (Y/N): ').upper() #memastikan konfirmasi yang dimasukkan berupa huruf kapital
            if pilihan_menu_awal in ['Y', 'N']: #looping untuk memastikan user hanya memasukkan pilihan yang valid, Y atau N
                return str(pilihan_menu_awal)
            else:
                raise Exception('Perintah yang Anda masukkan salah') 
            break
        except Exception as exc:
          print(exc)
  
#konfigurasi menu utama aplikasi Super Cashier
menu_utama = """
----------------------------------------------
Daftar pilihan transaksi:
1. Lihat isi keranjang belanja
2. Tambah item baru ke keranjang belanja
3. Update nama item di keranjang belanja
4. Update jumlah item di keranjang belanja
5. Update harga item di keranjang belanja
6. Hapus salah satu item di keranjang belanja
7. Hapus keranjang belanja
8. Keranjang belanja sudah sesuai
9. Cek ketentuan diskon
10. Batalkan transaksi
----------------------------------------------"""


#konfigurasi ketentuan diskon
'''format memasukkan data diskon:
[0, 0], [minimal belanja 1, diskon 1], [minimal belanja 2, diskon 2] '''
ref_diskon = [[0, 0], [200_000, 0.05], [300_000, 0.08], [500_000, 0.1]]
keterangan_diskon = ['Min Belanja' , 'Nilai Diskon (%)'] #label tabel diskon


#menampilkan tabel diskon
def tabel_diskon():
    ref_diskon2 = ref_diskon.copy() #mengcopy list ref_diskon dan memastikan sumber data awal tidak mengalami perubahan
    for i in range(len(ref_diskon)):
        ref_diskon2[i] = ref_diskon[i].copy() #mengcopy isi dari nested list ref_diskon sehingga isi ref_diskon tidak berubah
        for j in range(len(ref_diskon2[i])):
            ref_diskon2[i][1] = "{:.2%}".format(ref_diskon[i][1]) #mengubah format diskon menjadi persentase
    print(pd.DataFrame(ref_diskon2, columns = keterangan_diskon)) #menyajikan daftar diskon menjadi tabel
```

```python
class Transaksi:
  def __init__(self):
      id = str(random.randint(1,9999))
      self.id_transaksi = id_transaksi = str('ID-'+str(datetime.now(pytz.timezone('Asia/Jakarta')).strftime("%Y%m%d"))+(('0'*(4-len(id))+id)))
      self.keranjang = {'Ayam': [2, 25000], 'Daging': [2, 65000], 'Ikan': [3, 22500]}


  # teks konfirmasi akhir transaksi
  konfirmasi_transaksi = """
  ------------------------
  Konfirmasi transaksi:
  1. Check-out keranjang
  2. Kembali ke menu utama
  ------------------------"""


  # teks konfirmasi akhir transaksi
  konfirmasi_batal = """
  ------------------------
  Konfirmasi transaksi:
  1. Batalkan transaksi
  2. Kembali ke menu utama
  ------------------------"""


  def pilih_menu(self):
      pilihan_menu = int(input('masukan pilihan Anda (1-10): '))
      return pilihan_menu


  def pilih_konfirmasi(self):
      pilihan_konfirmasi = int(input('masukan pilihan Anda (1-2): '))
      return pilihan_konfirmasi


  def konfirmasi_tambah_item(self):
      while True:
          pilihan_konfirmasi = input('Apakah Anda ingin menambahkan item baru? (Y/N): ').title()
          try:
              if pilihan_konfirmasi not in ['Y', 'N']:
                  raise Exception('pilihan Anda tidak valid, ketik Y atau N')
              else:
                  return pilihan_konfirmasi
                  break
          except Exception as exc:
              print(exc)


  def tampilkan_keranjang(self):
      kolom = ['Nama Barang', 'Jumlah Barang', 'Harga per item', 'Subtotal Harga'] #setting standar label pada tabel
      item = []
      for i in self.keranjang.keys():
          subtotal_harga = self.keranjang[i][0]*self.keranjang[i][1] #menghitung subtotal harga per line item
          item.append([i, self.keranjang[i][0], self.keranjang[i][1], subtotal_harga])
      df = pd.DataFrame(item, columns = kolom) #menyajikan keranjang dalam bentuk tabel
      return df


  #menghitung total nilai belanja sebelum diskon
  def total_nilai_belanja(self):
      total_harga = 0
      for i in self.keranjang:
          total_harga += (self.keranjang[i][0]*self.keranjang[i][1])
      return total_harga


  #mencari diskon yang berlaku untuk total nilai pembelanjaan
  def cari_diskon(self):
      urut = []
      for i in range(len(ref_diskon)):
          if self.total_nilai_belanja() >= ref_diskon[i][0]:
              urut.append(i)
      return ref_diskon[len(urut)-1][1] #menggunakan panjang list untuk mencari applicable discount dari referensi diskon


  #menghitung nilai bersih pembelanjaan setelah diskon
  def net_total_price(self):
      net_price = self.total_nilai_belanja()*(1-self.cari_diskon())
      return net_price


  #menambahkan item ke keranjang
  def add_item(self):
      while True:
          item_name = input('nama barang: ').title()
          try: 
            if item_name == "" or item_name == " ": #memastikan user mengisi field ini 
                raise NameError('Nama barang harus diisi')
            elif item_name not in self.keranjang.keys():
                while True:
                    try:
                        jumlah = int(input('jumlah barang: '))
                        if jumlah < 0: #memastikan user hanya memasukkan nilai positif
                            raise Exception(f'Jumlah {item_name} tidak boleh negatif')
                        else:
                            break
                    except ValueError: #memastikan user hanya memasukkan numerik
                        print('Terdapat kesalahan, harap hanya memasukkan angka')
                    except Exception as exc:
                        print(exc)
                while True:
                    try:
                        harga = int(input('harga barang: Rp'))
                        if harga < 0: #memastikan user hanya memasukkan nilai positif
                            raise Exception(f'Harga {item_name} tidak boleh negatif')
                        else:
                            break
                    except ValueError: #memastikan user hanya memasukkan numerik
                        print('Terdapat kesalahan, harap hanya memasukkan angka')
                    except Exception as exc:
                        print(exc)
                self.keranjang.update({item_name: [jumlah, harga]}) #menambahkan item ke keranjang
                print(f'item {item_name}x{jumlah} @Rp{harga} telah ditambahkan ke keranjang')
                break

            else:
                print(f'{item_name} sudah ada di keranjang Anda') #mencari apakah nama barang sudah ada di keranjang
                print(f'Berikut adalah rincian barang tersebut [Nama, Qty, Harga/item]: {item_name}, {self.keranjang[item_name][0]}, {self.keranjang[item_name][1]}')
                while True: #looping input jumlah barang
                    try: #input hanya angka atau nuruf N
                        inputA = input('Masukkan jumlah untuk ditambahkan ke keranjang, atau "N" untuk membatalkan penambahan item: ')
                        if inputA.upper() == 'N': #input N artinya user tidak mengubah jumlah item
                            break
                        elif int(inputA) >= 0:
                            self.keranjang[item_name][0] += int(inputA) #update jumlah barang
                            break
                        else: #memastikan user tidak bisa memasukkan angka negatif
                            raise Exception('Nilai yang Anda masukkan tidak valid, hanya masukkan Angka positif atau N')
                    except Exception as exc:
                        print(exc)

                print(f'Harga {item_name} saat ini adalah {self.keranjang[item_name][1]}.')
                while True: #looping input harga barang
                    try: #input hanya angka atau huruf N (N untuk membatalkan)
                        inputB = input(f'Masukkan harga/per item yang baru untuk {item_name}, atau "N" jika tidak ingin mengubah harga barang: ')
                        if inputB.upper() == 'N':
                            break
                        elif round(float(inputB)) >= 0:
                            self.keranjang[item_name][1] = round(float(inputB)) #update harga barang
                            break
                        else: #memastikan user tidak bisa memasukkan angka negatif
                            raise Exception('Nilai yang Anda masukkan tidak valid, hanya masukkan Angka positif atau N')
                    except Exception as exc:
                        print(exc)
                break

          except NameError as VE:
              print(VE)


  #fungsi untuk mengupdate nama item belanja
  def update_item_name(self):
      print(f'\nBarang di keranjang Anda: {list(self.keranjang.keys())}')
      item_name = input('nama barang yang akan diubah: ').title()
      if item_name not in self.keranjang.keys():
          print(f'{item_name} belum ada di keranjang belanja Anda')
      elif item_name in self.keranjang.keys():
          while True: #memastikan user memasukkan nama barang baru
              nama_baru = input('masukkan nama barang yang baru: ').title()
              try:
                  if nama_baru == "" or nama_baru == " ":
                      raise NameError('Nama barang harus diisi') #memastikan input nama barang diisi oleh user
                  elif nama_baru in self.keranjang.keys():
                      raise Exception(f'{nama_baru} sudah ada di keranjang belanja Anda, silahkan pilih barang lain')
                  else:
                      self.keranjang[nama_baru] = self.keranjang.pop(item_name)
                      break
              except NameError as NE:
                  print(NE)
              except Exception as Exc:
                  print(Exc)


  #fungsi untuk mengupdate harga item belanja
  def update_item_price(self):
      print(f'\nBarang di keranjang Anda: {list(self.keranjang.keys())}')
      item_name = input('nama barang yang akan diubah quantity-nya: ').title()
      if item_name in self.keranjang.keys():
          print(f'harga item {item_name} saat ini adalah {self.keranjang[item_name][1]}')
          while True:
              try: #memastikan user hanya dapat memasukkan angka
                  price_baru = int(input('masukkan harga barang yang baru: Rp'))
                  if price_baru < 0: #memastikan angka yang dimasukkan tidak negatif
                      print('Harga barang tidak boleh negatif')
                  else:
                      self.keranjang[item_name][1] = price_baru
                      print(f'harga item {item_name} telah diubah menjadi Rp{price_baru}')
                      break
              except ValueError:
                print('Terdapat kesalahan, harap hanya memasukkan angka positif')
      else:
          print(f'Item {item_name} tersebut tidak ada di keranjang')


  def update_item_qty(self):
      print(f'\nBarang di keranjang Anda: {list(self.keranjang.keys())}')
      item_name = input('nama barang yang akan diubah quantity-nya: ').title()
      if item_name in self.keranjang.keys():      
          print(f'Jumlah item {item_name} di keranjang saat ini adalah {self.keranjang[item_name][0]}')
          while True:
              try: #memastikan user hanya dapat memasukkan angka
                  qty_baru = int(input(f'masukkan jumlah baru untuk item {item_name}: '))
                  if qty_baru < 0: #memastikan angka yang dimasukkan tidak negatif
                      print('Jumlah barang tidak boleh negatif')
                  else:
                      self.keranjang[item_name][0] = qty_baru
                      print(f'Jumlah item {item_name} telah diubah menjadi {qty_baru}')
                      break
              except ValueError:
                  print('Terdapat kesalahan, harap hanya memasukkan angka positif')
      else:
          print(f'Item {item_name} tersebut tidak ada di keranjang')
  

  # fungsi menghapus salah satu item dari keranjang belanja
  def delete_item(self):
      print(f'\nBarang di keranjang Anda: {list(self.keranjang.keys())}')
      try:
          item_name = input('nama barang yang akan dihapus: ').title()
          del self.keranjang[item_name]
          print(f'item {item_name} telah dihapus dari keranjang')
      except: #jika nama item tidak ada di keranjang, program hanya menampilkan pesan
          print('Barang tersebut tidak ada di keranjang')


  ##fungsi untuk menghapus seluruh isi keranjang belanja
  def reset_transaction(self):
      while True: #mengonfirmasi sebelum user reset yang sesungguhnya
          try: #memastikan user hanya memasukkan Y atau N
              konfirm = input('Apakah Anda yakin akan menghapus seluruh isi keranjang belanja? (Y/N): ').upper()
              if konfirm not in ['Y', 'N']:
                  raise Exception('Mohon masukkan input Y/N')
              elif konfirm == 'Y':
                  self.keranjang.clear()
                  print('seluruh isi keranjang telah dihapus')
              else:
                  pass
              break
          except Exception as exc:
            print(exc)

  #looping pilihan transaksi utama: tambah item, hapus item, hapus keranjang, edit item
  def loop_transaksi(self):
      while True:
          try:
              print(f'\n{menu_utama}')
              i = self.pilih_menu()
              if i == 8 or i == 10:
                  return i
                  break #jika user memilih 8 (check out) atau 10 (batalkan transaksi), maka looping menu utama akan berakhir
              elif i == 2:
                  print(' \nAnda memilih untuk menambahkan item ke keranjang')
                  self.add_item()
              elif i == 3:
                  print(' \nAnda memilih untuk mengubah nama item di keranjang')
                  print(f'{self.tampilkan_keranjang()}')
                  self.update_item_name()
              elif i == 4:
                  print(' \nAnda memilih untuk mengubah jumlah item di keranjang')
                  print(f'{self.tampilkan_keranjang()}')
                  self.update_item_qty()
              elif i == 5:
                  print(' \nAnda memilih untuk mengubah harga item di keranjang')
                  print(f'{self.tampilkan_keranjang()}')
                  self.update_item_price()
              elif i == 6:
                  print(' \nAnda memilih untuk menghapus salag satu item di keranjang')
                  print(f'{self.tampilkan_keranjang()}')
                  self.delete_item()
              elif i == 7:
                  self.reset_transaction()
              elif i == 1:
                  print(f'\n{self.tampilkan_keranjang()}')
              elif i == 9:
                  print(f'\nBerikut adalah syarat dan ketentuan diskon yang berlaku:')
                  tabel_diskon()
              else:
                  raise Exception("Perintah yang Anda masukkan tidak valid")
          except ValueError:
              print("Perintah yang Anda masukkan tidak valid")
          except Exception as exc:
              print(exc)
```

```python
#fungsi utama transaksi
def jalankan_aplikasi_cashier():
    pembukaan()
    a = buat_transaksi()
    if a == 'N':
        print("Terima kasih telah menggunakan aplikasi Super Cashier")
    else:
        trx123 = Transaksi()
        print(f'No. Transaksi: {trx123.id_transaksi} \n\nSilakan masukkan item belanja Anda')
        trx123.add_item()
        while True:
            if trx123.konfirmasi_tambah_item() == 'Y':
                trx123.add_item()
            else:
                break

        while True:
            a = trx123.loop_transaksi()
            if a == 8:
                print(f" \nBerikut adalah isi keranjang Anda \n{trx123.tampilkan_keranjang()} \n\ntotal sementara keranjang Anda: \033[1mRp{trx123.total_nilai_belanja()}\033[0m")
                print(trx123.konfirmasi_transaksi)
                b = trx123.pilih_konfirmasi()
                try:
                    if b == 1:
                        print(f'\n{trx123.tampilkan_keranjang()}')
                        diskon = "{:.2%}".format(trx123.cari_diskon()) 
                        print("")
                        print(f'Total nilai belanja Anda adalah \033[1mRp{trx123.total_nilai_belanja()}\033[0m')
                        if trx123.cari_diskon() == 0:
                            print(f"Nilai belanja Anda belum memenuhi syarat untuk memperoleh diskon \nTotal yang harus Anda bayar adalah \033[1mRp{trx123.net_total_price()}\033[0m")
                        else:
                            print(f'\033[1mSelamat, Anda memperoleh diskon sebesar {diskon}\033[0m')
                            print(f'''Total yang harus Anda bayar adalah \033[1mRp{trx123.net_total_price()}\033[0m \nTotal penghematan Anda adalah \033[1mRp{trx123.total_nilai_belanja()-trx123.net_total_price()}\033[0m''')
                        break
                except:
                    continue

            else:
                print(trx123.konfirmasi_batal)
                c = trx123.pilih_konfirmasi()
                try:
                    if c == 1:
                      a = 'Transaksi telah dibatalkan'
                      print('-'*len(a))
                      print(a)
                      print('-'*len(a))
                      break
                except:
                    continue
               
```
